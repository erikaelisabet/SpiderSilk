```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
all_path = "C:/Users/46705/Documents/SpiderSilk/data/pre_filtering/all_annotated_lascID.csv" 


df = pd.read_csv(all_path, sep=',')
```

```python
df.iloc[4:,:][['OG0000050', 'OG0009245', 'OG0013923', 'OG0000171', 'OG0002778',
       'OG0007640', 'OG0000494', 'OG0000264', 'OG0006927']]
```

```python
data = df.iloc[4:,:][['OG0000050', 'OG0009245', 'OG0013923', 'OG0000171', 'OG0002778',
       'OG0007640', 'OG0000494', 'OG0000264', 'OG0006927']].astype(float).apply(lambda x: np.log2(x + 1))


selected_data = data.T

plt.figure(figsize=(10, 6))
sns.heatmap(selected_data, cmap='rocket_r', linewidths=.5)
plt.title('Heatmap of Expression Levels')
plt.xlabel('Samples')
plt.ylabel('Genes')
plt.show()
```
